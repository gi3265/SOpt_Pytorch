소프트맥스 분류(Softmax Classification)에 대해 알아본다.
핵심키워드
* 소프트맥스(Softmax)
* 크로스 엔트로피(Cross Entropy)

Softmax란? 부드럽게 max를 뽑아주는 함수. max가 아닌 값들에 0을 때려버리지는 않음

Cross_Entropy란? 두 개의 확률분포(P(x), Q(x))가 주어졌을 때 그 두 확률분포가 얼마나 비슷한지 나타내는 수치

위의 Cross Entropy 수식 해석: P(x) 확률분포에서 x만큼 샘플링 하여 Q(x)에 집어넣고 로그를 취한 것의 평균이 Cross Entropy의 값(두 확률분포가 얼마나 비슷한지 나타내는 수치)이다.
